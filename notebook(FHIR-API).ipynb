```python
import requests
import json
# from pprint import pprint

headers = {
    'Accept': 'application/json+fhir',
}

params = (
    ('date', '2017'),
    ('patient', '4704007')
)

response = requests.get('https://fhir-open.sandboxcerner.com/dstu2/0b8a0111-e8e6-4c26-a91c-5069cbc6b1ca/Appointment', headers=headers, params=params).json()
```
